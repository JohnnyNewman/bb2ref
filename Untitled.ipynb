```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
game_id = "202002110HOU"

url = "https://www.basketball-reference.com" + f"/boxscores/{game_id}.html"
```

```python
print(url)
```

```python
r = requests.get(url)
```

```python
soup = bs(r.content)
```

```python
print(soup.prettify())
```

```python
for table in soup.find_all("table", "stats_table"):
    print(table.prettify())
    break
```

```python
table
```

```python
for tag in table.find_all("th", "left"):
    print(tag.text, tag.attrs)
```

```python
for table in soup.find_all("table", "stats_table"):
    for tag in table.find_all("th", "left"):
        if tag.a:
            print(tag.text, tag.a.get("href")[11:-5]) #, tag.attrs)
            print(tag.next_sibling.text)
    print()
```
